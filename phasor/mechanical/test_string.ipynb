```python
sys = system.system.BGSystem(
    #F_AC = logspaced(100, 2000, 10000),
    F_AC = np.linspace(499.999, 500.001, 100000),
)
#sys.own.R1 = mechanical.DisplacementReadout(
#    terminal = sys.F1.pm_A,
#)
#sys.own.M1 = mechanical.Mass(mass_kg = 1)
sys.own.M1 = mechanical.DisplacementSource()
sys.own.M2 = mechanical.Mass(mass_kg = 0.01)
sys.own.M3 = mechanical.Mass(mass_kg = 0.01)
sys.own.R1 = mechanical.DisplacementReadout(
    terminal = sys.M2.pm_A,
)
sys.own.S1 = mechanical.String(
    tension_N = 400,
    mass_density_kg_m = .0004,
    length_m = 1,
    Loss = 1e-7,
)
ratio_heavy = 1e2
sys.own.S2 = mechanical.String(
    tension_N = 1 * 4 * ratio_heavy,
    mass_density_kg_m = ratio_heavy * .0004,
    length_m = .1,
    Loss = 1e-4,
)
sys.M1.pm_A.bond(
    sys.S1.pm_A,
)
sys.M2.pm_A.bond(
    sys.S1.pm_B
)
sys.M2.pm_A.bond(
    sys.S2.pm_B
)
sys.M3.pm_A.bond(
    sys.S2.pm_A
)


sys.own.RAC1= readouts.ACReadoutLG(
    #portAct = sys.S1.pm_A.o,
    #portSense= sys.S1.pm_A.i,
    portAct = sys.S1.cable.pm_A.o,
    portSense= sys.S1.cable.pm_A.i,
)

sys.own.RAC2= readouts.ACReadout(
    portN = sys.R1.d.o,
    portD = sys.M1.d.i,
)

readoutI =  sys.RAC2

Fb = mplfigB(Nrows=2)
Fb.ax0.loglog(readoutI.F_Hz.val, abs(readoutI.AC_sensitivity))
Fb.ax1.semilogx(readoutI.F_Hz.val, np.angle(readoutI.AC_sensitivity, deg = True))
Fb.ax0.set_xscale('linear')
Fb.ax1.set_xscale('linear')

#readoutI.AC_sensitivity
readoutI =  sys.RAC1

Fb = mplfigB(Nrows=2)
Fb.ax0.loglog(readoutI.F_Hz.val, abs(readoutI.OLG))
Fb.ax0.loglog(readoutI.F_Hz.val, abs(readoutI.CLG))
Fb.ax1.semilogx(readoutI.F_Hz.val, np.angle(-readoutI.OLG, deg = True))
Fb.ax1.semilogx(readoutI.F_Hz.val, np.angle(readoutI.CLG, deg = True))
Fb.ax0.set_xscale('linear')
Fb.ax1.set_xscale('linear')

Fb = mplfigB(Nrows=1)
Fb.ax0.semilogy(readoutI.F_Hz.val, 1 - abs(readoutI.OLG))
Fb.ax0.set_xscale('linear')
print(max(abs(readoutI.CLG)), min(abs(readoutI.CLG)))
#readoutI.AC_sensitivity
```
